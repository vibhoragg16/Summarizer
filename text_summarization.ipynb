```python
##Prompt Template Text Summ

from langchain.chains import LLMChain
from langchain import PromptTemplate

generictemplate = """ 
Write a summary of the following speech:
Speech:{speech}
Translate the precise summary to {language}
"""

prompt = PromptTemplate(
    input_variable=['speech', 'language'],
    template = generictemplate
)

prompt

 
```

```python
complete_prompt = prompt.format(speech=speech, language="French")
complete_prompt
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
summary = llm_chain.run({'speech':speech,'language':'French'})
summary
```

# StuffDocumentChain TEXT SUMM

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
template = """Write a concise and short summary of the following speech
speech : {text}
"""

prompt = PromptTemplate(input_variable=['text'], template = template)
```

```python
from langchain.chains.summarize import load_summarize_chain

chain = load_summarize_chain(llm, chain_type='stuff', prompt=prompt, verbose=True)
output_summary = chain.run(docs)
output_summary
```

## Map Reduce Summarize Large Docs

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
final_document = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100).split_documents(docs)
final_document
```

```python
chunks_prompts = """ 
Please summarize the below speech:
Speech: '{text}'
Summary:
"""

map_prompt_template = PromptTemplate(input_variable=['text'],
                                     template=chunks_prompts)
```

```python
final_prompt = """ 
Provide the fianl summary of the entire speech with these important points.
Add a Motivational title, Start the precise summary with an introduction and provide the summary in number points for speech.
speech : {text}
"""

final_prompt_template = PromptTemplate(input_variable=['text'], template=final_prompt)
final_prompt_template
```

```python
summary_chain = load_summarize_chain(
    llm = llm,
    chain_type="map_reduce",
    map_prompt = map_prompt_template,
    combine_prompt = final_prompt_template,
    verbose = True
)

output = summary_chain.run(final_document)
output
```

## Refine Chain for Summarisation


```python
chain = load_summarize_chain(
    llm = llm, 
    chain_type="refine",
    verbose=True
)

output_summary=chain.run(final_document)
output_summary
```
